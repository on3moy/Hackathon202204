```python
document_list = ['Raw_Data/LA_NP_2015.csv','Raw_Data/OC_NP_2015.csv','Raw_Data/Rv_NP_2015.csv', 'Raw_Data/LA_NP_2016.csv','Raw_Data/OC_NP_2016.csv','Raw_Data/Rv_NP_2016.csv','Raw_Data/LA_NP_2017.csv','Raw_Data/OC_NP_2017.csv','Raw_Data/Rv_NP_2017.csv',
                 'Raw_Data/LA_NP_2018.csv','Raw_Data/OC_NP_2018.csv','Raw_Data/Rv_NP_2018.csv',
                 'Raw_Data/LA_NP_2019.csv','Raw_Data/OC_NP_2019.csv','Raw_Data/Rv_NP_2019.csv'
                 ]
year_list = [2015,2016,2017,2018,2019]
```

```python
df = pd.DataFrame()
for year in year_list:
    for excel in document_list:
        df_temp = pd.read_csv(excel, skiprows=4)
        if str(year) in excel:
            df_temp['year'] = year
            df = df.append(df_temp)


```
